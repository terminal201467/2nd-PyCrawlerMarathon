### ① 印出最新文章的「作者」「標題」「時間」

```python
import requests
from bs4 import BeautifulSoup

r = requests.get("https://www.ptt.cc/bbs/NBA/index.html","html5lib")
rt = r.text
soup = BeautifulSoup(rt)
soup
```

```python
#for d in soup.find_all(class_="title"):
title = [d.text.replace("\t","").replace("\n","") for d in soup.find_all(class_="title")]
date = [d.text for d in soup.find_all(class_="date")]
author = [ d.text for d in soup.find_all(class_="author")]
import pandas as pd
df = pd.DataFrame(columns = ["作者","標題","時間"])
df.作者 = author 
df.標題 = title
df.時間 = date
df
```

### ② 印出第一頁所有文章的「作者」「標題」「時間」

```python
import requests
from bs4 import BeautifulSoup
r = requests.get("https://www.ptt.cc/bbs/NBA/index1.html","html5lib")
rt = r.text
soup = BeautifulSoup(rt)
soup

author = [d.text for d in soup.find_all(class_="author")]
title = [d.text.replace("\t","").replace("\n","") for d in soup.find_all(class_="title")]
date = [d.text for d in soup.find_all(class_="date")]
import pandas as pd
df2 = pd.DataFrame(columns = ["作者","標題","時間"])
df2.作者 = author
df2.標題 = title
df2.時間 = date
df2
```

### ③ 試著爬爬看其他版的文章

```python
import requests as r
from bs4 import BeautifulSoup as b
res = r.get("https://www.ptt.cc/bbs/index.html")
rt = res.text
soup = b(rt)

title_manderin = [d.text.replace("\t", "").replace("\n","") for d in soup.find_all(class_="board-class")]
hit = [d.text for d in soup.find_all(class_="board-nuser")]
boardname_english = [d.text.replace("\t","").replace("\n","") for d in soup.find_all(class_="board-name")]
boardtitle = [d.text.replace("\t","").replace("\n","") for d in soup.find_all(class_="board-title")]

import pandas as pd
df = pd.DataFrame(columns = ["版類","英文分類名稱","版名","點擊率"])
df.版類 = title_manderin
df.英文分類名稱 = boardname_english
df.點擊率 = hit
df.版名 = boardtitle
df.head(10)
```
