```python
from bs4 import BeautifulSoup

soup = BeautifulSoup(r.text, 'html.parser')
table = soup.find('table', class_='TABLE_G')
print(table)
```

```python
d = {}
for tr in table.find_all("tr")[1:]:
    #print(tr)
    for i ,td in enumerate(tr.find_all("td")):
        if len(tr.find_all("td")) == 5: #如果五項的
            if i == 0:
               obs = td.text
               d.setdefault(obs,{})
            if i == 2:
               date = td.text
            if i == 3:
               value = td.text
        if len(tr.find_all("td")) == 3: #如果是三項的
            if i == 0:
                date = td.text
            if i == 1:
                value = td.text
                d[obs][date] = value
print(d)
```
