## 1. 取出今天所有的新聞

```python
import requests as r
from bs4 import BeautifulSoup as b
from selenium import webdriver as w

br = w.Chrome(executable_path = "C:/Users/user/chromedriver.exe")
br.get("https://www.ettoday.net/news/news-list.htm")
```

```python
from datetime import datetime , timedelta
date_time = datetime.now() - timedelta(1)
date_time = date_time.strftime("%Y%M%D")
print("取得時間:",date_time)
```

```python
from selenium import webdriver as w
import time
while True:
    br.execute_script("window.scrollTo(0, 100000);")
    time.sleep(2)
    html_source = br.page_source
    soup = b(html_source,"html5lib")
    print(soup.find_all(class_="part_list_2")[-1].find_all("h3")[-1].find(class_="date").text)

    if date_time in soup.find_all(class_="part_list_2")[-1].find_all('h3')[-1].find(class_="date").text:
        print("-----stop-----")
        break
```

## 2. 取出現在時間兩小時內的新聞

```python
from datetime import datetime ,timedelta

two_hours_ago = datetime.now() - timedelta(hours = 2)

news = []
for d in soup.find("div",attrs={"class":"part_list_2"}).find_all("h3"):
    if date_time.strptime(d.find(class_="date").text,"%Y/%m/%d %H:%M") < two_hours_ago:
        break
    news.append({
        "時間" : d.find(class_="date").text,
        "標題" : d.find_all("a")[-1].text
    })     

import pandas as pd
pd.DataFrame(news)
```

## 3. 根據範例，取出三天前下午三點到五點的新聞

```python
from datetime import datetime ,timedelta
three_day_period = datetime.now() - timedelta(days = 3) #引進今天減掉三天前的時間
three_day_period_month = three_day_period.strftime("%m") #月份
three_day_period_days = three_day_period.strftime("%d") #日期
three_day_1500 = datetime.strptime(three_day_period.strftime("%Y/%m/%d") + " 15","%Y/%m/%d %H") 
three_day_1700 = datetime.strptime(three_day_period.strftime("%Y/%m/%d") + " 17","%Y/%m/%d %H")

print("三天前的現在:",three_day_period) #三天前的現在
print("取得月份:",three_day_period_month)
print("取得日期",three_day_period_days)
print("1500時間:",three_day_1500)
print("1700時間:",three_day_1700)
```

```python
from selenium.webdriver.support.ui import Select

selM = Select(br.find_element_by_id("selM"))
selM.select_by_value(three_day_period_month)

selD = Select(br.find_element_by_id("selD"))
selD.select_by_value(three_day_period_days)

br.find_element_by_id("button").click()
```

```python
from bs4 import BeautifulSoup
import time
while True:
    br.execute_script("window.scrollTo(0,100000);")
    time.sleep(1)
    html_source = br.page_source
    soup = BeautifulSoup(html_source,"html5lib") 
    new_date = soup.find_all(class_="part_list_2")[-1].find_all("h3")[-1].find(class_="date").text
    print(new_date)
    if datetime.strptime(new_date,"%Y/%m/%d %H:%M")< three_day_1700:
        print("----stop----")
        break
```

```python
news = []

for d in soup.find(class_="part_list_2").find_all('h3'):
    # 比兩小時前早的話就結束
    if datetime.strptime(d.find(class_="date").text, '%Y/%m/%d %H:%M') < three_day_1500:
        print('==== STOP ====')
        break
    
    if datetime.strptime(d.find(class_="date").text, '%Y/%m/%d %H:%M') < three_day_1700:
        news.append({
            '時間': d.find(class_="date").text,
            '標題': d.find_all('a')[-1].text
        })

import pandas as pd
pd.DataFrame(news)  
```
