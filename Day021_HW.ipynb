## 作業目標

根據範例：

1. 取出今天所有的新聞
2. 取出現在時間兩小時內的新聞
3. 根據範例，取出三天前下午三點到五點的新聞

## 1. 取出今天所有的新聞

```python
import requests as r
from bs4 import BeautifulSoup as b
url = "https://www.ettoday.net/news/news-list.htm"
soup = b(r.get(url).text)

for d in soup.find("div",attrs={"class":"part_list_2"}).find_all("h3"):
    print(d.text)
```

## 2. 取出現在時間兩小時內的新聞

```python
from bs4 import BeautifulSoup as b
from selenium import webdriver as w

br = w.Chrome(executable_path = r"C:\Users\user\chromedriver")
br.get("https://www.ettoday.net/news/news-list.htm")
```

## 3. 根據範例，取出三天前下午三點到五點的新聞

```python
import time
for i in range(10):
    time.sleep(2)
    br.execute_script("window.scrollTo(0, 10000);")
```

```python
html_source = br.page_source
soup = b(html_source,"html5")

for d in soup.find("div",attrs={"class":"part_list_2"}).find_all("h3"):
    print(d.text)
br.quit()
```
